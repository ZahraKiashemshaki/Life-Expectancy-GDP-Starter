```python
df1= df.rename({"Life expectancy at birth (years)":"LEABY"}, axis= 'columns')
```

```python
plt.plot(df1.GDP, df1.Year, linestyle= '--', marker= 's', color= 'Green')
plt.xlabel('GDP')
plt.ylabel('Year')
plt.title('Relationship between GDP and years')
```

```python
plt.plot(df1.Country, df1.GDP, linestyle= ':', marker= '*', color= 'Red')
plt.xlabel('Country')
plt.ylabel('GDP')
plt.title('Relationship between Countries and GDP')
```

```python
plt.plot(df1.Country, df1.LEABY, marker= 'o', color= 'Grey')
plt.xlabel('Country')
plt.ylabel('LEABY')
plt.title('Relationship between Countries and LEABY')
```
